```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading year 2000 movie data into dataframe
the2000s_df = pd.read_csv('/Users/Rashad/Documents/GitHub/Test Repo/Project 3/Project-3/Data/final_tmdb_data_2000.csv.gz')
the2000s_df
```

```python
# Loading year 2001 movie data into dataframe
the2001s_df = pd.read_csv('/Users/Rashad/Documents/GitHub/Test Repo/Project 3/Project-3/Data/final_tmdb_data_2001.csv.gz')
the2001s_df
```

```python
# Merging both movie years into a single dataframe.
mov_2000_2001 = pd.merge(the2000s_df, the2001s_df, how='outer')
```

```python
mov_2000_2001.head(2)
```

```python
bud_rev = mov_2000_2001[ (mov_2000_2001['Movie_Budget'] > 0) & (mov_2000_2001['Movie_Revenue'] > 0) ]
bud_rev
```

```python
fig = sns.barplot(data=bud_rev, x='imdb_id', y='Movie_Budget');
fig.set(xlabel = "Movies", ylabel="Budget in Millions")
plt.title("Movie Budgets")
```

```python
bud_rev['MPAA_Rating'].value_counts().plot(kind='bar')
plt.title("Number of Movie Ratings")
```
